Introduction and Motivation:

As one of the most basic experiments in electricity and magnetism, the use of a statically charged hairbrush to curve water's path demonstrates a few key items about electric forces: namely, that electric forces can act at macroscopic distances to macroscopically affect the trajectory of macroscopic objects; that electric dipoles are especially susceptible to electric fields; and that neutral electric dipoles are only attracted, never repulsed, by electrically charged objects. Thus, it is near and dear to all learners of E & M.

When electric dipoles find themselves in an electric field, the most critical question is how orthogonal the dipole moment's orientation is compared to the electric field's orientation. If the dipole moment is pointing a different direction than the electric field, then the electric field will exert a torque on the dipole until its orientation matches that of the electric field.

If electric dipoles are in a non-uniform electric field, then, as there is energy stored in the electric field, they find themselves in a potential energy gradient. Thus, they experience a force in the direction of decreasing potential energy. As the electrical potential energy of a dipole decreases with increased electric field magnitude (see Setup), this means that the force will always be in the direction of greater electric field magnitude, which will always be in the direction of the nearest charge. This is why neutral electric dipoles always experience a force toward charges regardless of the charge's polarity.$^1$

Background:

In order to approximate the, the hairbrush will be approximated as a 1-dimensional line of electrical potential and the resulting potential throughout the rest of space will be calculated using the method of relaxation. 
	$V_{point}=(V_{left}+V_{right}+ V_{top}+V_{bottom})/4$

From this potential throughout space, the electric field throughout space will be found.

Because
	$E = \nabla V$,
we can approximate the electric field in the following way. (Please excuse the weird LaTex issues.)

    $E_r_x=(V{left}-{Vright})/2$

    $E_r_y=(V_{up}-V_{down})/2$

    $E= <E_r_x, E_r_y>$

Electric dipoles experience forces and torques in the presence of electric fields; the magnitudes of these can be given by the field’s interaction with the electric dipole moment.

	$U_dipole= -\vec p * \vec E$
	
Because 
$F = -\nabla U$,
we can treat the force on the electric dipole as the vector sum of its components.

	$\frac{dU}{dx}=\frac{d(-p*E)}{dx}= -(\frac{dp}{dx} *E + \frac{dE}{dx} *p)$
    
Because the dipole moment will not change in magnitude, only direction,
$\frac{dU}{dx}=\frac{d(-p*E)}{dx}= -\frac{dE}{dx} *p$. 

So,
	$F_x = -\frac{dU}{dx}= \frac{dE}{dx} *p
    
This process will be repeated to find the force vector on each dipole in the x and y directions. In the y-direction, there will also be the gravitational force acting on each molecule, causing it to accelerate downward. This will be summed with the y-component of the electric force on each dipole to find its acceleration.

Finally, the acceleration vector of each dipole will be used at each timestep, using Euler’s method, to update the water molecule’s velocity and position, such that its trajectory throughout space can be plotted$^3$.

For simplicity's sake, and because modeling dipole-dipole interactions in the stream is not a goal of this project, the stream will be modeled as one molecule.

Next, we establish the grid that our scalar potential function and electric field will reside in. These parameters have been tweaked a lot in order to find a balance between simulation fidelity and the feasibility of performing the method of relaxation of too large of a grid in a reasonable amount of time.

The printing of the shape of xGrid is a sanity check.

```python
#Establish x-y grid on which simulation will take place
maxX = 0.4
minX = 0.2
maxY = 1
minY = 0

xRes = 0.2/(1000)
yRes = 1/(1000)

x = np.arange(minX, maxX, xRes)
y = np.arange(minY, maxY, yRes)

xGrid, yGrid = np.meshgrid(x, y)
print(np.shape(xGrid))
```

In order to make the method of relaxation successful, we define a method for reestablishing the value of the scalar potential field in a region where it is known (i.e our hairbrush). Therefore, this function takes in the potential function to be acted upon, the bottom of the charged hairbrush region in meters, the top of the charged hairbrush region in meters, and the x-position of the line of hairbrush potential in meters. (hb is short for hairbrush).

In order to allow for the x-y coordinates of the hairbrush to be passed as coordinates in space, the x and y values are multiplied by how many elements are in the x and y arange spaces respectively to transform their float coordinate values into the corresponding index in V.

The potential per spot is difficult to know as the circumstances of the phenomenon are difficult to measure. However, as the experiment relies primarily on the electric field and its derivative, I decided to pick 100 V as the uniform potential across the hairbrush as this number, although probably quite unrealistically high, would allow for the electric field to hopefully be substantive.

```python
def establishInitialPotentialConditions(V,hbNadir, hbZenith, hbPos):
    v = V
    #Convert hairbrush boundaries from meters to indices
    hbXCoord = (int)((hbPos-0.2)/xRes)
    hbNadirYCoord = (int)(hbNadir * (1/yRes))
    hbZenithYCoord = (int)(hbZenith * (1/yRes))
    potentialPerSpot = 100
    v[hbNadirYCoord:hbZenithYCoord, hbXCoord] = potentialPerSpot
    return v
```

Now, we perform the method of relaxation by taking the average of the 4 potential values surrounding each point on the grid and then reestablishing our boundary conditions.

```python
def relax(V,hbNadir, hbZenith, hbPos): 
    iteratedV = (np.roll(V,-1, axis=0) + np.roll(V,1, axis=0) + np.roll(V,-1, axis=1) + np.roll(V,1, axis=1)) / 4
    resetV = establishInitialPotentialConditions(iteratedV,hbNadir, hbZenith, hbPos)    
    return resetV
```

Now, we iteratively perform that process to get a solved output grid that discretely describes our scalar potential function throughout space.

```python
def solveGrid(V, hbNadir, hbZenith, hbPos, iterCount = 20000):
    iterations = np.arange(iterCount)
    v = V
    for inum in iterations:
        v = relax(v,hbNadir, hbZenith, hbPos)
    return v
```

Now, we use the approximation noted in setup to find the gradient of our potential function, the electric field, throughout the same discretized space.

```python
def eField(V):
    eX = - (np.roll(V,-1,axis = 1) - np.roll(V,1,axis = 1))/2
    eY = - (np.roll(V,-1,axis = 0) - np.roll(V,1,axis = 0))/2
    return eX, eY
```

In order to find the derivative of the electric field with respect to x and with respect to y, we again apply the gradient operator to our found electric field by performing the same process on the electric field as we did to the potential.

```python
def eFieldDeriv(eField):
    derivX = - (np.roll(eField[0], -1,axis = 1) - np.roll(eField[0], 1,axis = 1))/2
    derivY = - (np.roll(eField[1], -1,axis = 0) - np.roll(eField[1], 1,axis = 0))/2
    return derivX, derivY
    
```

Next, we define getter methods to allow us to obtain the actual numerical values of the electric field and its derivative at any arbitrary point on our grid. The x and y coordinates, passed in as metric values, are again converted into indices to extract the numbers from our grids.

```python
def geteField(x,y, eFieldX, eFieldY):
    xCoord = (int)((x-0.2)*1000)
    yCoord = (int)(y*1000)
    eX = eFieldX[yCoord][xCoord]
    eY = eFieldY[yCoord][xCoord]
    return np.array([eX, eY])
```

```python
def geteFieldDeriv(x,y, eFieldDerivX, eFieldDerivY):
    xCoord = (int)((x-0.2)*1000)
    yCoord = (int)(y*1000)
    eDX = eFieldDerivX[yCoord][xCoord]
    eDY = eFieldDerivY[yCoord][xCoord]
    localEFieldDeriv = np.array([eDX, eDY])
    return localEFieldDeriv
```

We now go about the business of defining a dipole. The main operations that occur within this class are the simulation methods at the bottom which use Euler's Method to update both the dipole's orientation and its position in space.

These numerical values are then paired with their values in time using the enumeration of the timesteps array.

The position of the water dipole was constrained to be within x=0.20 to x=0.40 with the hairbrush at x=0.30 because home experiments led me to conclude that the hairbrush's range of effectiveness was quite small, so I figured it was best to not waste time and resources letting its range be larger than this.

```python
class Dipole():
    def __init__(self, Efield, EfieldDeriv, pos = np.array([0.305,0.999]), orientation = (1/np.sqrt(2))*np.array([-1,1]), mass = 18.016/6.02e23, 
               momentum = np.array([0,0]), omega = 0, dipoleMoment = 1.8546*3.33564e-30, momentOfInertia = 2.6e-47, dt =0.0001):
        self.pos = pos
        self.mass = mass
        self.orientation = orientation
        self.dipoleMoment = dipoleMoment
        self.momentOfInertia = momentOfInertia
        self.momentum = momentum
        self.dt = dt
        self.g = 9.81
        self.omega = omega
        self.timesteps = np.arange(0, 0.45, self.dt)
        self.durationOfSim = (int)(0.45/self.dt)
        self.momentumArray = np.zeros((self.durationOfSim, 2))
        self.posArray = np.zeros((self.durationOfSim, 2))
        self.orientationArray = np.zeros((self.durationOfSim, 2))
        self.Efield = Efield
        self.EfieldDeriv = EfieldDeriv
    
        
    def forceOnDipole(self, inum):
        f_g = self.mass * self.g
        F_g = f_g * np.array([0, -1])
        roundedX = np.around(self.pos[0], 3)
        roundedY = np.around(self.pos[1], 3)
        localeFieldVariance = geteFieldDeriv(roundedX, roundedY, self.EfieldDeriv[0], self.EfieldDeriv[1])
        #print(localeFieldVariance)
        if(np.linalg.norm(localeFieldVariance) != 0):
            f_e = np.dot(localeFieldVariance, self.dipoleMoment*self.orientationArray[inum][:])*localeFieldVariance/np.linalg.norm(localeFieldVariance)
        else:
            f_e = np.array([0,0])
        net_f = F_g + f_e
        return net_f
        
    def angularAccelOnDipole(self, inum):
        roundedX = np.around(self.pos[0], 3)
        roundedY = np.around(self.pos[1], 3)
        
        efield = geteField(roundedX, roundedY, self.Efield[0], self.Efield[1])
        #print(efield)
        tau = np.cross(self.dipoleMoment*self.orientationArray[inum][:], efield)
        return tau/self.momentOfInertia
    
    def updateOrientation(self, inum):
        if(inum == 0):
            self.orientationArray[inum][:] = self.orientation
        if(inum == 450000):
            pass
        else:
            self.orientationArray[inum][:] = self.orientationArray[inum - 1][:]
            attitude = self.orientationArray[inum][:]
            if(attitude[0] == 0):
                orientationRadians = np.pi/2
            else:
                orientationRadians = np.arctan(attitude[1]/attitude[0])
            alpha = self.angularAccelOnDipole(inum)
            self.omega = self.omega + (np.linalg.norm(alpha)*self.dt)
            orientationRadians += self.omega*self.dt
            self.orientationArray[inum][:] = np.array([np.cos(orientationRadians), np.sin(orientationRadians)])

    def updatePosition(self,inum):
        if(inum == 0):
            self.momentumArray[inum][:] = np.array([0,0])
            self.posArray[inum][:] = self.pos
        if(inum == 450000):
            pass
        else:
            self.momentumArray[inum][:] = self.momentumArray[inum - 1][:]
            #print(self.forceOnDipole(inum))
            self.momentumArray[inum][:] = self.momentumArray[inum][:] + self.forceOnDipole(inum)*self.dt
            self.posArray[inum][:] = self.posArray[inum - 1][:]
            #print(self.posArray[inum][:])
            self.posArray[inum][:] = self.posArray[inum][:] + (self.momentumArray[inum][:]/self.mass)*self.dt
        return self.momentumArray[inum][:], self.posArray[inum][:]
    
    def executeWaterSim(self):
        self.orientationArray[0][:] = self.orientation
        self.momentumArray[0][:] = self.momentum
        self.posArray[0][:] = np.array([0.25, 0.999])
        #print(self.posArray[0][:])
        for knum, time in enumerate(self.timesteps):
            self.updateOrientation(knum)
            #print("Orientation: ", self.orientationArray[knum - 1][:])
            updateData = self.updatePosition(knum)
            self.momentumArray[knum][:] = updateData[0]
            #print("Velocity: ", self.momentumArray[knum][:]/self.mass)
            self.posArray[knum][:] = updateData[1]
            #print("Position: ", self.posArray[knum][:])
    def extractData(self):
        return self.momentumArray, self.posArray, self.orientationArray
```
